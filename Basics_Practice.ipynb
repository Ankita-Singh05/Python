```
student_grade = int(input("Please enter a grade:"))

if student_grade>=90:
  letter ="A"
elif student_grade>=80:
  letter ="B"
elif student_grade>=70:
  letter = "C"
elif student_grade>=60:
  letter = "D"
else:
  letter ="F"
print(letter)
```

```
student_grade = int(input("Please enter a grade:"))

if student_grade<60:
  letter ="F"
elif student_grade>=60 and student_grade<70:
  letter ="D"
elif student_grade>=70 and student_grade<80:
  letter = "C"
elif student_grade>=80 and student_grade<90:
  letter = "B"
else:
  letter ="A"
print(letter)
```
